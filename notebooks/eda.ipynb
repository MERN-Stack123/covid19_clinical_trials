### Univariate

```python
# ===== TOP 20 MOST COMMON CONDITIONS =====
if 'conditions' in df.columns:
    top_condition = df['conditions'].value_counts().head()

    # Conditions Distribution
    plt.figure(figsize=(10,4))
    sns.barplot(x=top_condition.values, y=top_condition.index)
    plt.title("Top 20 Most Common Conditions")
    plt.xlabel("Count")
    plt.ylabel("Condition")
    plt.savefig('../images/top10_most_common_conditions', dpi=300, bbox_inches='tight')
    plt.show()
```

```python
#Status Distribution
plt.figure(figsize=(10,4))
sns.countplot(y=df["status"], order=df["status"].value_counts().index)
plt.title("Distribution of Trial Status")
plt.savefig('../images/distribution_trials_status', dpi=300, bbox_inches='tight')
plt.show()
```

```python
#Phases Distributions
plt.figure(figsize=(10,4))
sns.countplot(y=df["phases"], order=df["phases"].value_counts().index)
plt.title("Trial Phases Distributions")
plt.savefig('../images/distribution_trials_phases', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Study Type Distribution
plt.figure(figsize=(10,4))
sns.countplot(y=df["study_type"], order=df["study_type"].value_counts().index)
plt.title("Study Type Distributions")
plt.savefig('../images/study_type_distriubution', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Gender Distribution
plt.figure(figsize=(10,4))
sns.countplot(y=df['gender'], order=df['gender'].value_counts().index)
plt.title("Gender Distributions")
plt.savefig('../images/gender_distribution', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Age Distribution
plt.figure(figsize=(10,4))
sns.countplot(y="age", data=df, order=df["age"].value_counts().index[:20])
plt.title("Age Distributions")
plt.savefig('../images/age_distribution', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.kdeplot(df["enrollment"], fill=True)
plt.title("Enrollment KDE")
plt.savefig('../images/enrollment_kde', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# cheking outliers
plt.figure(figsize=(6,4))
sns.boxplot(x=df["enrollment"])
plt.title("Enrollment Boxplot")
plt.savefig('../images/enrollment_boxplot', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Start Year Distributions
plt.figure(figsize=(10,4))
dataType = df["start_year"].astype(int)
sns.countplot(y=dataType, order=dataType.value_counts().index)
plt.title("Start Year Distributions")
plt.savefig('../images/start_year_distribution', dpi=300, bbox_inches='tight')
plt.show()
```

### Bivariate Charts

```python
# Status vs Enrollment (boxplot)
plt.figure(figsize=(25,6))
sns.scatterplot(x=df['status'], y=df['enrollment'])
plt.savefig('../images/status_enrollment', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Phase vs Enrollment
plt.figure(figsize=(25,6))
sns.boxplot(x='phases', y='enrollment', data=df)
plt.savefig('../images/phase_enrollment', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# ===== TOP 10 Country =====

if 'country' in df.columns:
    top_countries = df['country'].value_counts().head(10)
    
    plt.figure(figsize=(10,4))
    sns.barplot(x=top_countries.values, y=top_countries.index)
    plt.title("Top 10 Country")
    plt.xlabel("Country")
    plt.ylabel("Count")
    plt.savefig('../images/top10_countries', dpi=300, bbox_inches='tight')
    plt.show()
```

```python
plt.figure(figsize=(12,4))
g = sns.catplot(x='phases', hue='status', kind='count', data=df)
g.set_xticklabels(rotation=45)
plt.title("Phases vs Status")
plt.savefig('../images/phases_vs_status', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.lineplot(x=df['start_year'], y=df.groupby('start_year')['nct_number'].count())
plt.savefig('../images/start_year_trend', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Heatmap (correlation)
plt.figure(figsize=(10,4))
numeric_df = df.select_dtypes(include='number')
sns.heatmap(numeric_df.corr(), annot=True)
plt.title("\n Correlation Heatmap (Numeric Columns Only)\n")
plt.savefig('../images/correlation_heatmap', dpi=300, bbox_inches='tight')
plt.show()
```
