### Univariate

```python
# ===== TOP 20 MOST COMMON CONDITIONS =====
if 'conditions' in df.columns:
    top_condition = df['conditions'].value_counts().head()

    # Conditions Distribution
    plt.figure(figsize=(10,4))
    sns.barplot(x=top_condition.values, y=top_condition.index)
    plt.title("Top 20 Most Common Conditions")
    plt.xlabel("Count")
    plt.ylabel("Condition")
    plt.savefig('../images/top10_most_common_conditions', dpi=300, bbox_inches='tight')
    plt.show()
```

```python
# Gender Distribution
plt.figure(figsize=(10,4))
sns.countplot(y=df['gender'], order=df['gender'].value_counts().index)
plt.title("Gender Distributions")
plt.savefig('../images/gender_distribution', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Age Distribution
plt.figure(figsize=(10,4))
sns.countplot(y="age", data=df, order=df["age"].value_counts().index[:20])
plt.title("Age Distributions")
plt.savefig('../images/age_distribution', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# funding source distribution
plt.figure(figsize=(10,4))
df['funded_bys'].value_counts().plot(kind='bar')
plt.title("Trials by Funding Source")
plt.xlabel("Funding Source")
plt.ylabel("Count")
plt.savefig('../images/funding_source_distribution', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# study type distribution
plt.figure(figsize=(10,4))
df['study_type'].value_counts().plot(kind='bar')
plt.title("Study Type Distribution")
plt.xlabel("Study Type")
plt.ylabel("Count")
plt.savefig('../images/study_type_distriubution', dpi=300, bbox_inches='tight')
plt.show()
```

### Bivariate Charts

#### Monthly trial start trend

```python
df['YearMonth'] = df['start_year'].astype(str) + "-" + df['start_month'].astype(str)
monthly_trials = df.groupby('YearMonth').size()

monthly_trials.plot(figsize=(12,4))
plt.xticks(rotation=90)
plt.title("Monthly Clinical Trial Start Trend")
plt.savefig('../images/monthly_trial_trend', dpi=300, bbox_inches='tight')
plt.show()
```

#### Sponsor vs Enrollment size

```python
plt.figure(figsize=(10,4))
df.groupby('funded_bys')['enrollment'].mean().plot(kind='bar')
plt.title("Average Enrollment by Funding Source")
plt.xlabel("Funding Source")
plt.ylabel("Average Enrollment")
plt.savefig('../images/average_enrollment_by_funding_source', dpi=300, bbox_inches='tight')
plt.show()
```

#### Defining succuss & failure

```python
df['Outcome'] = df['status'].apply(lambda x: 'Success' if x == 'Completed' else 'Failure')

plt.figure(figsize=(10,4))
df['Outcome'].value_counts(normalize=True).plot(kind='bar')
plt.title("Success vs Failure Rate")
plt.xlabel("Outcome")
plt.ylabel("Proportion")
plt.savefig('../images/success_failure_rate', dpi=300, bbox_inches='tight')
plt.show()
```

#### Phase-wise Success rate

```python
plt.figure(figsize=(10, 2))
pd.crosstab(df['phases'], df['Outcome'], normalize='index').plot(kind='bar')
plt.title("Phase-wise Success Rate")
plt.xlabel("Phase")
plt.ylabel("Proportion")
plt.savefig('../images/phase_wise_success_rate', dpi=300, bbox_inches='tight')
plt.show()
```

#### Vaccine vs Drug-Based Trial Comparison

```python
df['Trial_Type'] = df['interventions'].apply(lambda x: 'Vaccine' if 'vaccine' in str(x).lower() else 'Drug')
plt.figure(figsize=(10,4))
df['Trial_Type'].value_counts().plot(kind='bar')
plt.title("Vaccine vs Drug Trials")
plt.xlabel("Trial Type")
plt.ylabel("Count")
plt.savefig('../images/vaccine_vs_drug_trials', dpi=300, bbox_inches='tight')
plt.show()
```

#### Success rate comparision

```python
plt.figure(figsize=(10,4))
pd.crosstab(df['Trial_Type'], df['Outcome'], normalize='index').plot(kind='bar')
plt.title("Success Rate: Vaccine vs Drug")
plt.xlabel("Trial Type")
plt.ylabel("Success Rate")
plt.savefig('../images/success_rate_vaccine_vs_drug', dpi=300, bbox_inches='tight')
plt.show()
```
